## LLM Summarization

Use Hugging Face Transformers to summarize the entries for each month

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('../data/cleaned/daily_sentences_cleaned.xlsx')
```

```python
from transformers import pipeline


summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
monthly_summaries = {}

for month, group in df.groupby('Month'):
    text = " ".join(group['Sentence'])
    if len(text.split()) > 500:  # Truncate for long months
        text = " ".join(text.split()[:500])
    summary = summarizer(text, max_length=60, min_length=25, do_sample=False)[0]['summary_text']
    monthly_summaries[str(month)] = summary

for m, s in monthly_summaries.items():
    print(f"{m}: {s}\n")
```
